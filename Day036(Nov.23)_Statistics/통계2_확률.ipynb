## 확률

- 사건이 일어날 확률
- 고전적 확률 정의
      실험결과가 N가지 일 때 사건 A가 m번 일어난 경우 P(A) = m/N (각 실험 가능성은 모두 같아야함)
      (예시) 주사위 {1,2,3,4,5,6} 하나의 숫자를 뽑을 확률은 모두 1/6으로 같다

- 경험적 확률 정의
      반복 횟수가 B일 경우, 무한 반복했을때 발생하는 사건 A의 횟수가 m이면, 사건 A의 확률은
      P(A) = lim m/N

- 공리적 확률 정의
      콜모고로프 공리적 정의 세가지
      0 <= P(A) <= 1
      P(S) = 1
      실험이 표본공간의 부분집합이면서 다른 사건과 교집합이 없을 때 각 실험의 확률을 더한 값과 합집합의 확률은 같다.
      예를 들면 동전 두개 던지기를 할때
      앞면 : H, 뒷면 : T
      표본공간 = S{HH, HT, TH, TT}
      앞이 나오는 사건(A{HH})와 뒷면이 나오는 사건(B{TT})는 교집합이 없다.
      P(AUB) =P({HH, TT}) =P({TT})+P({HH})
      P(AUB)= P(A)+P(B)


## 확률의 성질
- 공집합의 확률 0
- 여집합의 확률
      실험 A의 여집합은 1-P(A)

- 포함-배제 원리
      P(AUB) = P(A) + P(B) - P(AnB)

- 각 사건이 배반사상(event)일 경우
      P(A1 U A2 U A3 .... U An) = P(A1)+P(A2)....P(An)
      A가 B의 부분집합일 때 P(A) <= P(B)

## 조건부 확률
- 어떤 사건 B가 일어날 때(P(B)>0) 사건 A가 일어날 확률
$$P(A|B) = \frac{P(A \cap B)}{P(B)}$$

- 사건 A가 발생했을 때 사건 B가 발생할 확률
$$P(B|A) = \frac{P(A \cap B)}{P(B)}$$

$$P(A\cap B) = P(A|B)P(B) = P(B|A)P(A)$$

- 독립사건
$$P(A|B) = P(A)$$
$$P(B|A) = P(B)$$

## 베이즈 정리

- 분할사건
      사건 K개가 일어났을 때 사건 A1 ... Ak 모두  다 교집합이 없는 경우 분할 사건
      사건 K개를 모두 합집합했을 때 S가 되는 경우

$$ B = (B\cap {A_1}) U ... U (B\cap {A_k})$$
$$ P(B) =P(B\cap {A_1})+...+P(B\cap{A_k})$$

$$P({A_i}|B) = \frac{P(B|{A_i})P({A_i})}{P(B)} = \frac{P(B|{A_i})P({B_i})}{P(B|{A_1})P({A_1})+P(B|{A_2})P({A_2})+...+P(B|{A_k})P({A_k}))}$$

사전 확률을 사후확률로 추정하는 정리

## 확률 변수와 확률 분포

- 확률변수
      - 표본공간에서 정의된 실수값 함수
      - 확률 변수들의 통계량을 찾고
      - 확률 변수를 이용해서 모수 추정
      - 대문자 X로 표현

- 확률 분포
      확률 변수 값과 확률을 대응시키는 표, 그래프, 함수

- 이산 확률 변수
      Countably infinite 변수
      확률 질량 함수를 통해 확률을 얻을 수 있다.

    - 확률 질량 함수

  $$P(a<=X<=b) = Σf(x)$$


- 연속 확률 변수
      하나로 특정할 수 없고 셀수 없는 변수
    - 확률 밀도 함수
          확률 밀도 함수 f(x)는 연속확률 변수 X에 대한 확률에 분포를 나타내는 함수
          모든 x에 대해서 f(x) >= 0
          연속밀도함수(CDF), 확률밀도함수(PDF)
          확률 밀도 함수 면적의 합은 1
          특정 구간에서의 확률은 확률밀도 함수를 특정 구간에 대해 적분한 값이다.
          

## 정규 분포
- 정의
      확률 변수 X의 평균이 μ 이고 표준편차가 σ 인 정규분포는 
$$N(μ, {σ^2})$$

- 특징
      모든 정규분포는 좌우 대칭
      정규곡선의 위치는 평균에 의해서 결정된다.
      표본공간에 따라 평균과 표준편차는 다르지만 평균으로부터 k배 표준편차 이내에 확률 변수 X가 존재할 확률은 동일

## 표준 정규 분포

  - 확률 변수를 정규화 시켜서 표준편차 1이고 평균이 0인 정규분포로 만든 것

## 분산과 표준편차

- 분산 확률분포함수의 모양을 설명할 수 있다.
- 분산은 확률분포함수에 확률이 모여있는지 퍼져있는지를 알 수 있는 수치적 파라미터

$${σ^2} = Var[X] = E[{(X-\mu)^2}]$$

- 성질
      Var[X] >= 0
      Var[cX] = c^2Var[X]

      Var[X+Y] = Var[X] + Var[Y] + 2E[(X-m_x)(Y-m_y)]



- 표본평균의 분산과 표본의 분산의 관계
$$E[X] = E[\bar{X}]$$
$$Var[\bar{X}] = \frac{1}{N}Var[X]$$
