```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,100)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x-0), color='blue')           #색상 지정방법 1. 색상이름
plt.plot(x, np.sin(x-1), color='g')              #색상 지정방법 2. 색상약어(짧은 색상코드. RGB, CMYK)
plt.plot(x, np.sin(x-2), color='0.2')            #색상 지정방법 3. (0-1)사이에 회색조를 표현한다.
plt.plot(x, np.sin(x-3), color='#FFDD44')        #색상 지정방법 4. 헥스코드 - 16진수로 표현한 RGB : R자리(FF),G자리(DD),B자리(44)
plt.plot(x, np.sin(x-4), color=(1.0, 0.2, 0.3))  #색상 지정방법 5. RGB 값을 튜플로 표현 R자리(1),G자리(0.2),B자리(0.3)
plt.plot(x, np.sin(x-5), color='chartreuse')     #색상 지정방법---색상이름 번외
```

```python
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

```python
plt.plot(x, x+0, '-g')
plt.plot(x, x+1, '--b')
plt.plot(x, x+2, '-.r')
plt.plot(x, x+3, ':c')
```

```python
# 플롯 조정하기: 축 경계

plt.plot(x, np.sin(x))

plt.xlim(-1,11)     # x축 경계 조정
plt.ylim(-1.5,1.5)  # y축 경계 조정
```

```python
plt.plot(x, np.sin(x))

plt.xlim(10,0)     # x축 좌우 반전
plt.ylim(1.5,-1.5)  # y축 상하 반전
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5])
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight')  #경계에 꽉차도록
```

```python
# equal - 데이터 진위파악 할때 사용
# 실제 좌표계

plt.plot(x, np.sin(x))
plt.axis('equal')      
```

```python
# 플롯에 레이블 붙이기

plt.plot(x, np.sin(x))
plt.title('A sine Curve')
plt.xlabel('X')
plt.ylabel('sin(x)')
```

```python
plt.plot(x, np.sin(x), '-.b', label='y = sin(x)')
plt.plot(x, np.cos(x), '-r', label='y = cos(x)')
plt.axis('equal')

plt.legend()        #범례
```

```python
# 산점도의 출력

x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')      # 'o' = marker
```

```python
# Marker의 종류

rng = np.random.RandomState(0)   # Random 값을 만드는 numpy기능(시드 값임)
#'o' 동그란점
#',' 
#'x' x표시
#'+' +표시
#'v' 아래쪽화살표
#'<' 왼쪽 삼각형
#'>' 오른쪽 삼각형
#'s' 사각형
#'d' 마름모
for marker in ['o',',','x','+','v','<','>','s','d']:
  plt.plot(rng.rand(5), rng.rand(5), marker, label='marker={0}'.format(marker))
  plt.legend()
  plt.xlim(0,1.8)
```
