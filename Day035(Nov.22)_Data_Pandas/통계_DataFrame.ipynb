```python
## 실습

index=['철수','영희','길동','슬기']
columns = ['국어','영어','수학','과학']
data = {
'국어' : [80,90,70,50],
'영어' : [90,50,60,100],
'수학' : [70,100,50,70],
'과학' : [30,50,80,80],
}

df_p1 = pd.DataFrame(data, index=index, columns=columns)
df_p1
```

```python
df_p1['학생평균'] = df_p1.mean(axis=1)
```

```python
df_p1.loc['과목평균'] = df_p1.mean()
```

```python
df_p1
```

```python
(df_p1['국어']+df_p1['영어']+df_p1['수학']+df_p1['과학'])/4
```

```python
df_p1['학생평균'] = (df_p1['국어']+df_p1['영어']+df_p1['수학']+df_p1['과학'])/4
```

```python
df_p1
```

```python
(df_p1.loc['철수']+df_p1.loc['영희']+df_p1.loc['길동']+df_p1.loc['슬기'])/4
```

```python
# 학생 평균 컬럼 접근
df_p1.loc['과목평균'] = (df_p1.loc['철수']+df_p1.loc['영희']+df_p1.loc['길동']+df_p1.loc['슬기'])/4
```
