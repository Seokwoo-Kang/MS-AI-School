```python
import pandas as pd
```

```python
# 두개의 DF를 하나의 DF로 합치기
df1 = pd.DataFrame({
    '날짜' : [1101,1102,1103,1104,1105],
    '온도' : [20,18,17,14,5]
    
}, columns=['날짜','온도'])
```

```python
df1
```

```python
df2 = pd.DataFrame(data={
    '날짜' : [1101,1102,1103,1104,1109, 1120],
    '습도' : [10 ,20, 40, 100, 6, 50]
}, columns=['날짜', '습도'])
```

```python
df2
```

```python
pd.merge(df1,df2)
```

merge()의 경우 중복되는 열에 대해 합쳐진다
default 값이 how = 'inner'

```python
pd.merge(df1,df2,how='outer')
```

```python
# 온도를 기준으로 합치고 싶을때

pd.merge(df1,df2,how='left')
```

```python
# 습도를 기준으로 합치고 싶을때

pd.merge(df1,df2,how='right')
```

```python
# 날짜를 기준으로
pd.merge(df1,df2,how='right', on='날짜')
```

```python
df3 = pd.DataFrame({
    'date' : [1101,1102,1103,1104,1105],
    '온도' : [20,18,17,14,5]
    
}, columns=['date','온도'])
```

```python
df4 = pd.DataFrame(data={
    '날짜' : [1101,1102,1103,1104,1109, 1120],
    '습도' : [10 ,20, 40, 100, 6, 50]
}, columns=['날짜', '습도'])
```

```python
pd.merge(df3,df4, left_on='date', right_on='날짜')
```
